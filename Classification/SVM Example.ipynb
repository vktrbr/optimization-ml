```python
import torch
import pandas as pd
from algorithms import SVM
from sklearn.datasets import make_classification, make_blobs
from sklearn.model_selection import train_test_split

x, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, random_state=1)
# x, y = make_blobs(n_samples=10, n_features=2, centers=2, random_state=777)
x = torch.FloatTensor(x)
y = torch.FloatTensor(y) * 2 - 1
x = x / torch.abs(x).max()

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

## Train

```python
model = SVM(2)
model = model.fit(x_train, y_train, 5000)
```

## Test

```python
metrics = model.metrics_tab(x_test, y_test, metric='by_roc')
tab = pd.DataFrame({'-1': metrics['-1.0'], '1': metrics['1.0']}).T.round(3)
print(f'Accuracy: {metrics["accuracy"]: 0.2%} | Threshold: {metrics["threshold"]: 0.2f}')
tab
```
